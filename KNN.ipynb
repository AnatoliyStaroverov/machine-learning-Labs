```
import numpy as np
import matplotlib.pyplot as plt 
from matplotlib.colors import ListedColormap
from sklearn import neighbors, datasets

n_neighbors = 20

bc = datasets.load_breast_cancer()

X = bc.data[: , :2]
y = bc.target

cmap_light = ListedColormap(['#FFAAAA','#AAFFAA','#AAAAFF']) # background
cmap_bold = ListedColormap(['#FF0000','#00FF00','#0000FF']) # data points

clf = neighbors.KNeighborsClassifier(n_neighbors)
clf.fit(X,y)

# define grid 
x_min,x_max = X[:,0].min() -1 , X[:,0].max() + 1
y_min,y_max = X[:,0].min() -1 , X[:,0].max() + 1
xx,yy = np.meshgrid(np.arange(x_min,x_max,.02),np.arange(y_min,y_max, .02)) 

Z = clf.predict(np.c_[xx.ravel(), yy.ravel()]) # flatten to 1d
Z = Z.reshape(xx.shape)

plt.figure()
plt.pcolormesh(xx,yy,Z,cmap=cmap_light)

plt.scatter(X[:,0],X[:,1], c=y,cmap=cmap_bold,edgecolor='k',s=20)
plt.xlim(xx.min(),xx.max())
plt.ylim(yy.min(),yy.max())
plt.title("Binary classification (K = %i)" % (n_neighbors))

plt.show()
```
